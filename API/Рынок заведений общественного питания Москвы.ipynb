### Сетевое распространение

***Выясню какому типу объекта характерно сетевое распространение:***

```python
rest_data_object_chain = rest_data.query('chain == "да"').groupby('object_type').agg({'id': 'count'}).reset_index().rename(columns={'id':'chain_yes'})
rest_data_all_object_type = rest_data.groupby('object_type')['id'].count().reset_index().rename(columns={'id':'all_objects_in_category'})
rest_data_object_chain = rest_data_all_object_type.merge(rest_data_object_chain, on='object_type')
rest_data_object_chain['ratio'] = round(100*(rest_data_object_chain['chain_yes']/rest_data_object_chain['all_objects_in_category']), 1)
rest_data_object_chain = rest_data_object_chain.sort_values(by='ratio', ascending=False)
```

```python
fig = px.bar(rest_data_object_chain, x='object_type', y='ratio', color='object_type', 
            hover_data=['all_objects_in_category','chain_yes'], labels={'object_type':'Заведения'},
            color_discrete_sequence=px.colors.sequential.haline)
fig.update_layout(margin=dict(l=50, r=0, t=90, b=0), yaxis_title="Соотношение", xaxis_title=" ", 
                  plot_bgcolor="white", title='Сетевое распространение')
fig.update_xaxes(showticklabels=False)
fig.show('svg')
```

***Выделю в отдельный столбец информацию об улице из столбца address***

```python
all_street = ['вал', 'зона', 'массив', 'сад', 'сквер', 'проулок', 
          'разъезд', 'спуск', 'тракт', 'просека', 'ряд', 'ряды', 
          'проспект', 'площадь', 'улица', 'шоссе', 'бульвар', 
          'переулок', 'набережная', 'проезд', 'тупик', 'микрорайон',
          'территория', 'километр', 'аллея', 'просек', 'линия', 'квартал']

def street(address):
    chosen_address = address.split(',') 
    all_index = len(chosen_address) 
    for word in all_street:
        for index in range(all_index):
            try:
                if word in chosen_address[index]:
                    return chosen_address[index]
            except:
                return 'ошибка'
                
    
rest_data['street'] = rest_data['address'].apply(street)
display(rest_data.sample(5))
```

Дальше, согласно данной документации https://yandex.ru/dev/maps/geocoder/doc/desc/concepts/input_params.html
составлю заспрос на нахождение координат и района:

```python
api = '...'# укажите ваш API если есть. Если нет, то ниже я подготовила готовые таблицы

def location(address):
    try:
        URL = 'https://geocode-maps.yandex.ru/1.x/?geocode='+address+'&format=json&apikey='+api+'&kind=district'
        params = {"city":"Moscow"}
        response = requests.get(URL, params=params)
        response_parsed = json.loads(response.text)
        coordinates = response_parsed['response']['GeoObjectCollection']['featureMember'][0]['GeoObject']['Point']['pos']
        
        URL_district = 'https://geocode-maps.yandex.ru/1.x/?geocode='+coordinates+'&format=json&apikey='+api+'&kind=district'
        response_district = requests.get(URL_district, params=params)
        response_parsed_district = json.loads(response_district.text)
        district = response_parsed_district['response']['GeoObjectCollection']['featureMember'][0]['GeoObject']['metaDataProperty']['GeocoderMetaData']['AddressDetails']['Country']['AdministrativeArea']['Locality']['DependentLocality']['DependentLocality']['DependentLocalityName']
        return district
    except:
        return 'ошибка'
    

#test = location('город москва, ленинградский проспект, дом 76а')
#test 
```
